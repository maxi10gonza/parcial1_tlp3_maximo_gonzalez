# Importamos la pandas

```python
import pandas as pd
```

```python
df = pd.read_csv('votaciones.csv')
df.head(30)
```

```python
df.shape
```

```python
df.columns
```

```python
df.dtypes
```

```python
df.info()
```

```python
df.isnull().sum()
```

Eliminamos filas con valores nulos

```python
df = df.dropna()
```

```python
df.shape
```

```python
df.head(10)
```

Convertimos fecha_votacion al tipo datatime

```python
df['fecha_votacion'] = pd.to_datetime(df['fecha_votacion'], errors='coerce')
```

```python
df.head(10)
```

Colocamos en nombre, apellido y provincia mayuscula a las primeras letras

```python
df['nombre'] = df['nombre'].str.strip().str.lower().str.title()
df['apellido'] = df['apellido'].str.strip().str.lower().str.title()
df['provincia'] = df['provincia'].str.strip().str.lower().str.title()
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

Convertimos dni a tipo entero

```python
df['dni'] = df['dni'].astype(int)
```

```python
df.dtypes
```

Colocamos la columna 'voto' en mayusculas

```python
df['voto'] = df['voto'].str.strip().str.upper()
```

```python
df.head(10)
```

Eliminamos las filas con fechas NaT

```python
df = df.dropna(subset=['fecha_votacion'])
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.head(20)
```

Estadistica descriptiva

```python
df.describe()
```

Estadisticas agrupadas

```python
grupo = df.groupby(['provincia', 'voto'])
print(grupo['dni'].agg(['count','mean','min','max']))
```

Importamos matplotlib para la visualizacion

```python
import matplotlib.pyplot as plt
```

# Grafico de barras de los distintos tipos de votos (SÍ, NO, NULO)

```python
conteo = df.groupby(['provincia', 'voto']).size().unstack(fill_value=0)

conteo.plot(kind='bar', figsize=(10,6))

plt.title('Cantidad de votos por provincia')
plt.xlabel('Provincia')
plt.ylabel('Cantidad de votos')
plt.xticks(rotation=45)
plt.legend(title='voto')
plt.tight_layout()
plt.show
```

Exportamos a archivo SQlite

```python
import sqlite3
```

```python
conn = sqlite3.connect('votantes.db')

df.to_sql('personas', conn, index=False)
```
